1. 자료 가져오기 (pandas)

```
import pandas as pd   # pandas 를 pd 로 불러온다.

url = 'https://vincentarelbundock.github.io/Rdatasets/csv/HistData/Guerry.csv'
original_df = pd.read_csv(url)
original_df = pd.read_csv(url, index_col=0)
# `index_col=0`은 첫 열을 index로 지정
original_df.head()
```

```
original_df.tail()
```

```
original_df.dropna(inplace=True)
```

```
original_df.tail()
```

```
df = original_df[['Crime_prop', 'Literacy', 'Wealth', 'Distance']]
```

```
df.head()
```

2. OLS 회귀분석 (statsmodels)

```
import statsmodels.api as sm
```

```
import statsmodels.formula.api as smf
```

```
res = smf.ols(formula='Crime_prop ~ Literacy + Wealth + Distance', data=df).fit()
res
```

```
print(res.summary())
```

4. 예측 (scikit-learn)

```
from sklearn.linear_model import LinearRegression
```

```
model = LinearRegression()
```

```
x = df[['Literacy', 'Wealth', 'Distance']]  # 학습 자료
y = df['Crime_prop']  # 정답 자료
```

```
model.fit(x, y)
```

# 분석 데이터 변환
일일 데이터 사용량을 5개의 군집으 로 나누어 평활한다.

```
from sklearn.cluster import KMeans
```

```
# 클러스터 생성

kms=KMeans(n_clusters=5)
```

```
스마트폰데이터사용량=스마트폰.loc[
                    (
                        스마트폰['데이터사용량'] != 0
                    )
]
```

```
len(스마트폰데이터사용량)
```

```
kms.fit(스마트폰데이터사용량)
```

```
kms
```

```
kms.labels_
```

```
kms.predict(스마트폰데이터사용량)
```

```
center=kms.cluster_centers_
print(center)
```

```
스마트폰_cbind=pd.concat([스마트폰, pd.DataFrame(kms.cluster_centers_)], axis=1)
```

```
스마트폰_cbind.head()
```

```
스마트폰_cbind=pd.concat([스마트폰, pd.DataFrame(kms.labels_)], axis=1)
```

```
스마트폰_cbind.head()
```

```
스마트폰_cbind['cluster'] = kms.labels_
```

```
스마트폰_cbind.head()
```

```
# 위의 데이터를 시각화처리( seaborn 사용) - 각 클러스터별 다른 색으로 , 150개의 사이즈로 출력
import seaborn as sns

sns.lmplot( '연령' , '데이터사용량', data=스마트폰_cbind, fit_reg=False, scatter_kws={"s":150},hue="cluster" )
plt.title('K-Means exam')
```

```
스마트폰_cbind.to_csv("스마트폰_cbind.csv")
```
