## 시리즈 클래스

### 시리즈 생성

```python
# 인덱스 없이 시리즈 생성
S1=pd.Series(
    range(11,15)
)

S1
```

```python
# 인덱스 있는 상태에서 시리즈 생성
S2=pd.Series(
    [1,2,3,4],
    index=["봄","여름","가을","겨울"]
)

S2
```

*시리즈에* 대한 인덱스와 값 조회

```python
#S1
print(S1.index)
print(S1.values)
```

```python
#S2
print(S2.index)
print(S2.values)
```

시리즈 인덱싱

```python
# 봄과 겨울에 대한 평균
# 번호
print((S2[0]+S2[3])/2)
# 이름
print((S2["봄"]+S2["겨울"])/2)
```

시리즈에 대한 명명

```python
# 시리즈에 명명
S2.name="계절정보"

# 인덱스에 대한 명명
S2.index.name="계절"
# S2.values.name="계절코드" 에러남

S2
```

### 데이터의 갱신, 추가, 삭제

인덱싱을 이용하면 딕셔너리처럼 데이터를 갱신(update)하거나 추가(add)할 수 있다.

```python
S2["봄"] = 11
S2
```

데이터를 삭제할 때도 딕셔너리처럼 ``del`` 명령을 사용한다.

```python
del S2["봄"]
S2
```

```python
data={                
        "2015": [9904312, 3448737, 2890451, 2466052],
        "2010": [9631482, 3393191, 2632035, 2431774],
        "2005": [9762546, 3512547, 2517680, 2456016],
        "2000": [9853972, 3655437, 2466338, 2473990],
        "지역": ["수도권", "경상권", "수도권", "경상권"],
        "2010-2015 증가율": [0.0283, 0.0163, 0.0982, 0.0141]
    }
index=["서울","부산","인천","대구"]
columns=["지역","2000","2005","2010","2015","2010-2015 증가율"]

DF3=pd.DataFrame(data,index=index,columns=columns)

DF3
```

### 열 데이터의 갱신, 추가, 삭제

데이터프레임은 열 시리즈의 딕셔너리으로 볼 수 있으므로 열 단위로 데이터를 갱신하거나 추가, 삭제할 수 있다.

```python
# "2010-2015 증가율"이라는 이름의 열 추가
DF3["2010-2015 증가율"] = DF3["2010-2015 증가율"] * 100
DF3
```

```python
# "2005-2010 증가율"이라는 이름의 열 추가
DF3["2005-2010 증가율"] = ((DF3["2010"] - DF3["2005"]) / DF3["2005"] * 100).round(2)
DF3
```

```python
# "2010-2015 증가율"이라는 이름의 열 삭제
del DF3["2010-2015 증가율"]
DF3
```

# 틱 설정

```python
import numpy as np

X = np.linspace(-np.pi, np.pi, 256)
C = np.cos(X)
plt.title("x축과 y축의 tick label 설정")
plt.plot(X, C)
plt.xticks([-np.pi, -np.pi / 2, 0, np.pi / 2, np.pi])
plt.yticks([-1, 0, +1])
plt.show()
```

```python
X = np.linspace(-np.pi, np.pi, 256)
C = np.cos(X)
plt.title("LaTeX, 문자열로 tick label 정의")
plt.plot(X, C)
plt.xticks([-np.pi, -np.pi / 2, 0, np.pi / 2, np.pi],
           [r'$-\pi$', r'$-\pi/2$', r'$0$', r'$+\pi/2$', r'$+\pi$'])
plt.yticks([-1, 0, 1], ["Low", "Zero", "High"])
plt.show()
```

# 그리드 설정

```python
X = np.linspace(-np.pi, np.pi, 256)
C = np.cos(X)
plt.title("Grid 제거")
plt.plot(X, C)
plt.xticks([-np.pi, -np.pi / 2, 0, np.pi / 2, np.pi],
           [r'$-\pi$', r'$-\pi/2$', r'$0$', r'$+\pi/2$', r'$+\pi$'])
plt.yticks([-1, 0, 1], ["Low", "Zero", "High"])
plt.grid(False)
plt.show()
```

# 범례

```python
X = np.linspace(-np.pi, np.pi, 256)
C, S = np.cos(X), np.sin(X)
plt.title("legend를 표시한 플롯")
plt.plot(X, C, ls="--", label="cosine")
plt.plot(X, S, ls=":", label="sine")
plt.legend(loc=2)
plt.show()
```

# x축, y축 라벨, 타이틀

```python
X = np.linspace(-np.pi, np.pi, 256)
C, S = np.cos(X), np.sin(X)
plt.plot(X, C, label="cosine")
plt.xlabel("time")
plt.ylabel("amplitude")
plt.title("Cosine Plot")
plt.show()
```

Axes 객체와 subplot 명령

```python
x1 = np.linspace(0.0, 5.0)
x2 = np.linspace(0.0, 2.0)
y1 = np.cos(2 * np.pi * x1) * np.exp(-x1)
y2 = np.cos(2 * np.pi * x2)

ax1 = plt.subplot(2, 1, 1)
plt.plot(x1, y1, 'yo-')
plt.title('A tale of 2 subplots')
plt.ylabel('Damped oscillation')

ax2 = plt.subplot(2, 1, 2)
plt.plot(x2, y2, 'r.-')
plt.xlabel('time (s)')
plt.ylabel('Undamped')

plt.tight_layout()
plt.show()
```

```python
np.random.seed(0)

plt.subplot(221)
plt.plot(np.random.rand(5))
plt.title("axes 1")

plt.subplot(222)
plt.plot(np.random.rand(5))
plt.title("axes 2")

plt.subplot(223)
plt.plot(np.random.rand(5))
plt.title("axes 3")

plt.subplot(224)
plt.plot(np.random.rand(5))
plt.title("axes 4")

plt.tight_layout()
plt.show()
```

```python
fig, axes = plt.subplots(2, 2)

np.random.seed(0)
axes[0, 0].plot(np.random.rand(5))
axes[0, 0].set_title("axes 1")
axes[0, 1].plot(np.random.rand(5))
axes[0, 1].set_title("axes 2")
axes[1, 0].plot(np.random.rand(5))
axes[1, 0].set_title("axes 3")
axes[1, 1].plot(np.random.rand(5))
axes[1, 1].set_title("axes 4")

plt.tight_layout()
plt.show()
```

Axis 객체와 축

```python
fig, ax0 = plt.subplots()
ax1 = ax0.twinx()
ax0.set_title("2개의 y축 한 figure에서 사용하기")
ax0.plot([10, 5, 2, 9, 7], 'r-', label="y0")
ax0.set_ylabel("y0")
ax0.grid(False)
ax1.plot([100, 200, 220, 180, 120], 'g:', label="y1")
ax1.set_ylabel("y1")
ax1.grid(False)
ax0.set_xlabel("공유되는 x축")
plt.show()
```
