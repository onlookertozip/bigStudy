```python
import numpy as np
import scipy as sp

# 소수점 이하 자릿수 3
%precision 3
```

### 1변량 데이터 관리

```python
fish_data = np.array([2,3,3,4,4,4,4,5,5,6])
fish_data
```

### 합계와 샘플사이즈

```python
# 합계
sp.sum(fish_data)
```

```python
# 샘플사이즈
len(fish_data)
```

### 평균값

```python
N = len(fish_data)
sum_value = sp.sum(fish_data)
mu = sum_value / N
mu
```

```python
sp.mean(fish_data)
```

### 표본분산

```python
sigma_2_sample = sp.sum((fish_data - mu) ** 2) / N
sigma_2_sample
```

```python
fish_data
```

```python
fish_data - mu
```

```python
(fish_data - mu) ** 2
```

```python
sp.sum((fish_data - mu) ** 2)
```

```python
sp.var(fish_data, ddof = 0)
```

ddof : int, 선택 사항
델타 자유도를 의미합니다. 계산에 사용되는 제수는 N-ddof이며, 여기서 N은 요소 수를 나타냅니다. 기본적으로 ddof는 0입니다.모분산은 ddof = 0, 표본분산은 ddof = 1
넘파이는 ddof = 0이 디폴트, 판다스는 ddof = 1 이 디폴트

### 불편분산

```python
sigma_2 = sp.sum((fish_data - mu) ** 2) / (N - 1)
sigma_2
```

```python
sp.var(fish_data, ddof = 1)
```

### 표준편차

```python
sigma = sp.sqrt(sigma_2)
sigma
```

```python
sp.std(fish_data, ddof = 1)
```

### 표준화

```python
fish_data - mu
```

```python
sp.mean(fish_data - mu)
```

```python
fish_data / sigma
```

```python
sp.std(fish_data / sigma, ddof = 1)
```

```python
standard = (fish_data - mu) / sigma
standard
```

```python
sp.mean(standard)
```

```python
sp.std(standard, ddof = 1)
```

#기술통계 예-와인 품질 데이터 셋

```python
#!/usr/bin/env python3
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import statsmodels.api as sm
import statsmodels.formula.api as smf
from statsmodels.formula.api import ols, glm


# Read the data set into a pandas DataFrame
wine = pd.read_csv('winequality-both.csv', sep=',', header=0)
wine.columns = wine.columns.str.replace(' ', '_')
print(wine.head())

# Display descriptive statistics for all variables
print(wine.describe())

# Identify unique values
print(sorted(wine.quality.unique()))

# Calculate value frequencies
print(wine.quality.value_counts())

# Display descriptive statistics for quality by wine type
print(wine.groupby('type')[['alcohol']].describe().unstack('type'))

# Calculate specific quantiles
print(wine.groupby('type')[['quality']].quantile([0.25, 0.75]).unstack('type'))
```

# 기술통계 예-고객 데이터 셋

```python
#!/usr/bin/env python3
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import statsmodels.api as sm
import statsmodels.formula.api as smf

# Read the data set into a pandas DataFrame
churn = pd.read_csv('customer.csv', sep=',', header=0)

churn.columns = [heading.lower() for heading in \
churn.columns.str.replace(' ', '_').str.replace("\'", "").str.strip('?')]

churn['churn01'] = np.where(churn['churn'] == 'True.', 1., 0.)
print(churn.head())
print(churn.describe())


# Calculate descriptive statistics for grouped data
print(churn.groupby(['churn'])[['day_charge', 'eve_charge', 'night_charge', 'intl_charge', 'account_length', 'custserv_calls']].agg(['count', 'mean', 'std']))

# Specify different statistics for different variables
print(churn.groupby(['churn']).agg({'day_charge' : ['mean', 'std'], 
				'eve_charge' : ['mean', 'std'],
				'night_charge' : ['mean', 'std'],
				'intl_charge' : ['mean', 'std'],
				'account_length' : ['count', 'min', 'max'],
				'custserv_calls' : ['count', 'min', 'max']}))
```
