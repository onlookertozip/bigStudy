## 슬라이싱

```python
g = list(range(20))
g
```

```python
g[0:5]
```

이 때 앞의 0이라는 숫자는 생략할 수 있다.

```python
g[:5]
```

만약 11 번째에서 15번째까지의 자료라면 다음과 같다.

```python
g[10:15]
```

## 역 인덱싱(음수 인덱싱)

```python
g[-1]
```

음수 인덱스를 사용할 때도 슬라이싱을 할 수 있다. 두 번째 인덱스가 생략되면 끝까지를 의미한다(-1을 쓸 때와 다르다).

```python
g[-5:-2]
```

```python
g[-3:]
```

```python
g[-3:-1]
```

## 딕셔너리 자료형의 반복

```python
for k in x:
    print(k)
```

```python
for k in x.keys():
    print(k)
```

딕셔너리 자료형의 값 반복

```python
x.values()
```

```python
for v in x.values():
    print(v)
```

딕셔너리 자료형의 키와 값의 쌍 반복

```python
for k, v in x.items():
    print("key [%s] => value [%d]" % (k, v))
```
